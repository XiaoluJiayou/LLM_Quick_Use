```python
import pandas as pd
from pipeline_unit2ixl import UniT2IXLPipeline
```

```python
data = pd.read_csv("jade_benchmark_zh.csv")
```

```python
data.head()
len(data)
```

```python
for i in range(10):
    print(data.iloc[i]["提示词"])
```

```python
pipeline = UniT2IXLPipeline.from_pretrained("/root/autodl-tmp/UnicomAI/UniT2IXL")
# pipeline.load_lora_weights("lora.safetensors")
pipeline = pipeline.to("cuda:0")
```

```python
import os
from pathlib import Path
base_path = "img_result"
for i in range(10, 227):
    prompt = data.iloc[i]["提示词"]
    image = pipeline(prompt=prompt,guidance_scale=7.5,target_size=(1024,1024)).images[0]
    file_path = "{}.png".format(i)
    full_file_path = os.path.join(base_path, file_path)
    image.save(full_file_path)
```
